```python
import requests
import multiprocessing as mp
import boto3
import json
from boto3.dynamodb.conditions import Key
import pandas as pd

# set up db
db = boto3.resource('dynamodb', region_name = 'us-east-2')
table = db.Table('TryInmates')
```

```python
!aws dynamodb scan --table-name TryInmates --region us-east-2 --output json > ./inmates.json
```

```python
with open('./inmates.json') as fp:
    s = fp.read()
    j = json.loads(s)
```

```python
raw_df = pd.DataFrame(j['Items'])
df = pd.DataFrame()
```

```python
list(raw_df.columns)
```

```python
for col in list(raw_df.columns):
    df[col] = raw_df[col].apply(lambda e: e['S'])
```

```python
anonymized_df = df[['sex', 'age',  'race','region', 'projRelDate', 
                     'actRelDate',  'releaseCode', 'faclName','faclType', 'faclCode', 'faclURL']]
```
